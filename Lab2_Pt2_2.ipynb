```python
# Import statements
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

Use pandas to read the data located in the subfolder data.

```python
# Load dataset
df = pd.read_csv('/content/drive/MyDrive/ADEWALE DATA ANALYSIS /water_usage.csv')
df
```

```python
df.shape
```

Use a pie chart to visualize the water usage. Highlight one usage of your choice using the explode parameter. Show the percentages for each slice and add a title.

```python
# Create figure
plt.figure(figsize=(4,4), dpi=300)
# Create pie plot
x = df.Percentage
labels = df.Usage

plt.pie(x,labels=labels, explode = [0,0,0,0,0.05,0], startangle=90, counterclock=False, autopct='%.1f%%')
# Add title
plt.title('Water Usage')
# Show plot
plt.tight_layout()
plt.show()
```
