## Water usage

```python
# Import statements
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

Use pandas to read the data located in the subfolder data.

```python
# Load dataset
df = pd.read_csv('../Data/water_usage.csv')
df
```

```python
df.shape
```

We use a pie chart to visualize the water usage, highlighting toilet usage using the explode parameter.

```python
# Create figure
plt.figure(figsize=(4,4), dpi=300)
# Create pie plot
x = df.Percentage
labels = df.Usage

plt.pie(x,labels=labels, explode = [0,0,0,0,0.05,0], startangle=90, counterclock=False, autopct='%.1f%%')
# Add title
plt.title('Water Usage')
# Show plot
plt.tight_layout()
plt.show()
```
