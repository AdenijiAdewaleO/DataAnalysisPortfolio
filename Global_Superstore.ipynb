From the above statistical description of the data, we see that the profit column has negative values (min of -6599.978). This means that some of the transactions made lossess (negative profits). Our analysis here is therefore to find out how the transactions fare in losses against the total sales.

## AN INTERESTING GRAPH

For our one interesting graph, we look at the total sales and profits by country.

```python
#Group the dataset by country and sum the profits and sales for each country
country_profits = df.groupby('Country').sum()['Profit']
country_sales = df.groupby('Country').sum()['Sales']
```

```python
country_df = pd.merge(country_profits, country_sales, on='Country', how='inner')
country_df.reset_index(inplace=True)
```

```python
country_df.sort_values(by='Profit', ascending=True, inplace=True)
country_viz = country_df.head(20)
country_viz
```

```python
# bar plot of sales and profit by country on the same axis

plt.bar(country_viz['Country'], country_viz.Sales, label='Sales')
plt.bar(country_viz['Country'], country_viz.Profit, label='Loss')
plt.xlabel('Country')
plt.xticks(rotation='vertical')
plt.title('Sales and Loss by top 20 countries with the most losses')
plt.legend()
plt.show()
```

## RESULT ANALYSIS
From the above bar graph, we can see how the losses made in the top 20 countries(rated by losses) compare with the total sales made in these countries. The following key points are highlighted:

- Turkey has the highest total loss among all the countries
- Countries like Nigeria, Lithuania, Kazakhstan and Zimbabwe are seen to have lossess that surpasses the total sales.

There would need for further analysis to determine what are the contributing factors to the losses and why some countries have more losses than sales
